```python
import json
from collections import defaultdict
import statistics

# path to your JSON file
file_path = "/content/bin_dev.json"

# load the JSON file (assuming it's a list of items)
with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# dictionary: lemma_id -> set of gloss_ids
lemma_gloss_map = defaultdict(set)

for item in data:
    lemma_id = item["lemma_id"]
    gloss_id = item["gloss_id"]
    lemma_gloss_map[lemma_id].add(gloss_id)

# convert sets to counts
gloss_counts = {lemma: len(glosses) for lemma, glosses in lemma_gloss_map.items()}

# collect all counts in a list
counts = list(gloss_counts.values())

# compute statistics
mean_val = statistics.mean(counts)
median_val = statistics.median(counts)
try:
    mode_val = statistics.mode(counts)  # most common value
except statistics.StatisticsError:
    mode_val = "No unique mode"

min_val = min(counts)
max_val = max(counts)

# print per-lemma stats
print("Lemma statistics (number of glosses per lemma):")
for lemma, count in gloss_counts.items():
    print(f"{lemma}: {count} glosses")

# print overall stats
print("\nOverall statistics:")
print(f"Total lemmas: {len(gloss_counts)}")
print(f"Mean (average): {mean_val:.2f}")
print(f"Median: {median_val}")
print(f"Mode: {mode_val}")
print(f"Minimum glosses per lemma: {min_val}")
print(f"Maximum glosses per lemma: {max_val}")
```
